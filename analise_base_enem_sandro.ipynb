```python
from scipy import stats
import pandas as pd

# ── 1. Filtro de baixa renda ──────────────────────────────────────────────────
faixas_baixa_renda = ['Nenhuma Renda', '<= 998.00', '998.00 até 1497.00'] # no caso defini aqui até R$ 1497,00

df_baixa_renda = base_enem[base_enem['renda_mensal_familiar'].isin(faixas_baixa_renda)].copy()

print(f"Total de candidatos de baixa renda: {len(df_baixa_renda):,}")
print(df_baixa_renda['internet_na_residencia'].value_counts())
```

```python
# ── 2. Separação dos grupos ───────────────────────────────────────────────────
sem_internet = df_baixa_renda[df_baixa_renda['internet_na_residencia'] == 'A']
com_internet = df_baixa_renda[df_baixa_renda['internet_na_residencia'] == 'B']

print(f"\nCom internet:  {len(com_internet):,} candidatos")
print(f"Sem internet:  {len(sem_internet):,} candidatos")
```

```python
# ── 3. Estatísticas descritivas ───────────────────────────────────────────────
notas = ['nota_redacao', 'nota_mt']

for nota in notas:
    print(f"\n{'='*50}")
    print(f"  {nota.upper()}")
    print(f"{'='*50}")
    print(f"  Com internet → média: {com_internet[nota].mean():.2f} | mediana: {com_internet[nota].median():.2f}")
    print(f"  Sem internet → média: {sem_internet[nota].mean():.2f} | mediana: {sem_internet[nota].median():.2f}")
```

```python
# ── 4. Teste de Mann-Whitney U (unilateral) ───────────────────────────────────
# alternative='greater' → testa se com_internet > sem_internet  (alinhado com H1)

alpha = 0.05 # definição do erro, 0,05 sendo padrão

print("\n" + "="*60)
print("  RESULTADOS DO TESTE DE MANN-WHITNEY U")
print("="*60)

for nota in notas:
    grupo_com = com_internet[nota].dropna()
    grupo_sem = sem_internet[nota].dropna()

    stat, p_valor = stats.mannwhitneyu(grupo_com, grupo_sem, alternative='greater')

    print(f"\n  {nota.upper()}")
    print(f"     Estatística U : {stat:.2f}")
    print(f"     p-valor       : {p_valor:.6f}")

    if p_valor < alpha:
        print(f"     ✅ Rejeita H0 — evidência de que internet melhora a nota (α={alpha})")
    else:
        print(f"     ❌ Não rejeita H0 — sem evidência suficiente (α={alpha})")

    # rejeita h0, ou seja nao concorda com a hipótese: O acesso à internet não influencia as notas de estudantes de baixa renda.
```

```python
# ── 5. Visualização comparativa ───────────────────────────────────────────────
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Distribuição das Notas — Baixa Renda\nCom vs. Sem Internet', fontsize=14, fontweight='bold')

titulos = {'nota_redacao': 'Redação', 'nota_mt': 'Matemática'}

for ax, nota in zip(axes, notas):
    dados_plot = pd.DataFrame({
        'Nota': pd.concat([com_internet[nota], sem_internet[nota]], ignore_index=True),
        'Grupo': ['Com Internet'] * len(com_internet[nota]) + ['Sem Internet'] * len(sem_internet[nota])
    }).dropna()

    sns.boxplot(
        data=dados_plot,
        x='Grupo', y='Nota',
        palette={'Com Internet': '#2196F3', 'Sem Internet': '#FF7043'},
        ax=ax,
        showfliers=False   # remove outliers extremos para melhor visualização
    )

    # linha da mediana de cada grupo
    mediana_com = com_internet[nota].median()
    mediana_sem = sem_internet[nota].median()
    ax.set_title(f'{titulos[nota]}\n(mediana: com={mediana_com:.0f} | sem={mediana_sem:.0f})', fontsize=11)
    ax.set_xlabel('')
    ax.set_ylabel('Nota')

plt.tight_layout()
plt.savefig('hipotese_internet_baixa_renda.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ══════════════════════════════════════════════════════════════════
#  BÔNUS — CORRELAÇÃO: PEARSON vs SPEARMAN
#  Heatmap comparativo com múltiplas variáveis
# ══════════════════════════════════════════════════════════════════

# ── 1. Converter renda para numérico (ponto médio de cada faixa) ──
mapa_renda = {
    'Nenhuma Renda'          : 0,
    '<= 998.00'              : 499,
    '998.00 até 1497.00'     : 1247,
    '1497.00 até 1996.00'    : 1746,
    '1996.00 até 2495.00'    : 2245,
    '2495.00 até 2994.00'    : 2744,
    '2994.00 até 3992.00'    : 3493,
    '3992.00 até 4990.00'    : 4491,
    '4990.00 até 5988.00'    : 5489,
    '5988.00 até 6986.00'    : 6487,
    '6986.00 até 7984.00'    : 7485,
    '7984.00 até 8982.00'    : 8483,
    '8982.00 até 9980.00'    : 9481,
    '9980.00 até 11976.00'   : 10978,
    '11976.00 até 14970.00'  : 13473,
    '14970.00 até 19960.00'  : 17465,
    '>= 19960.00'            : 19960
}

base_enem['renda_numerica'] = base_enem['renda_mensal_familiar'].map(mapa_renda)
```

```python
# ── 2. Subconjunto com as variáveis selecionadas ──────────────────
colunas = [
    'idade', 'pessoas_na_residencia',
    'nota_lc', 'nota_ch', 'nota_ct', 'nota_mt', 'nota_redacao',
    'renda_numerica'
]

# Labels mais legíveis para o heatmap
labels = {
    'idade'                 : 'Idade',
    'pessoas_na_residencia' : 'Pessoas\nResidência',
    'nota_lc'               : 'Linguagens',
    'nota_ch'               : 'C. Humanas',
    'nota_ct'               : 'C. Natureza',
    'nota_mt'               : 'Matemática',
    'nota_redacao'          : 'Redação',
    'renda_numerica'        : 'Renda'
}

df_corr = base_enem[colunas].dropna()
print(f"Registros utilizados: {len(df_corr):,}")
```

```python
# ── 3. Calcular matrizes de correlação ────────────────────────────
matrix_pearson  = df_corr.rename(columns=labels).corr(method='pearson')
matrix_spearman = df_corr.rename(columns=labels).corr(method='spearman')

# Máscara para ocultar triângulo superior (evita redundância)
import numpy as np
mask = np.triu(np.ones_like(matrix_pearson, dtype=bool))
```

```python
# ── 4. Heatmaps comparativos ──────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle(
    'Correlação de Pearson vs Spearman\n'
    'Comparação da robustez entre os dois métodos',
    fontsize=14, fontweight='bold'
)

heatmap_params = dict(
    mask=mask,
    annot=True,
    fmt='.2f',
    vmin=-1, vmax=1,
    center=0,
    cmap='coolwarm',
    square=True,
    linewidths=0.5,
    annot_kws={'size': 9}
)

sns.heatmap(matrix_pearson,  ax=axes[0], **heatmap_params)
axes[0].set_title(
    'Pearson\n(assume normalidade e linearidade)',
    fontsize=11, pad=10
)

sns.heatmap(matrix_spearman, ax=axes[1], **heatmap_params)
axes[1].set_title(
    'Spearman\n(baseado em rankings — robusto a outliers)',
    fontsize=11, pad=10
)

plt.tight_layout()
plt.savefig('bonus_heatmap_pearson_vs_spearman.png', dpi=150, bbox_inches='tight')
plt.show()

# correlação considerada forte: C. Humanas por Linguagens. Ou seja normalmente quem vai bem em um, também acompanha o nível do outro.
```

```python
# ── 5. Tabela de diferenças entre os dois métodos ─────────────────
diff = (matrix_spearman - matrix_pearson).abs()

# Pega os pares únicos (triângulo inferior, sem diagonal)
pares = []
cols = list(matrix_pearson.columns)
for i in range(len(cols)):
    for j in range(i):
        pares.append({
            'Variável A'  : cols[i],
            'Variável B'  : cols[j],
            'Pearson'     : round(matrix_pearson.iloc[i, j], 4),
            'Spearman'    : round(matrix_spearman.iloc[i, j], 4),
            'Diferença'   : round(diff.iloc[i, j], 4)
        })

df_diff = pd.DataFrame(pares).sort_values('Diferença', ascending=False)

print("\n Top 10 pares com maior diferença entre Pearson e Spearman:")
print(df_diff.head(10).to_string(index=False))
```

```python
# POR QUE PEARSON E SPEARMAN DIVERGEM TÃO POUCO NESSA BASE?
#
# As notas do ENEM são calculadas pelo TRI (Teoria de Resposta ao
# Item), que por design gera distribuições próximas do normal —
# simétricas e sem distorções extremas. Isso faz com que Pearson,
# que assume normalidade, funcione bem para as notas.
#
# A maior diferença aparece justamente em RENDA, que é assimétrica
# à direita (muitos ganham pouco, poucos ganham muito) — exatamente
# onde Pearson perde robustez e Spearman se sai melhor.
#
# Conclusão: quando as variáveis são aproximadamente normais,
# Pearson e Spearman convergem. A diferença aparece em variáveis
# assimétricas como renda, onde Spearman captura melhor a relação.
```
